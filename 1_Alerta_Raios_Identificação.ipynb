```python
#Importação de Pacotes
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

import warnings
warnings.filterwarnings("ignore")
```

```python
#Função leitura de data e hora
dateparse = lambda x: datetime.strptime(x, '%Y %m %d %H %M %S')
```

```python
# Abertura da base de dados

dataset = pd.read_csv('/content/drive/MyDrive/Alerta_Raios_Análise/chuva_dados/1_tempestade_26_11_2011_0.txt', 
                    sep='\s+', 
                    header=None,
                    parse_dates={ 'horario': [0, 1, 2, 3, 4, 5]},
                    date_parser=dateparse,
                    names=['ano', 'mes', 'dia', 'hora', 'm', 's', 'E'])
#dataset.info()
```

## **2) Visualização - Gráfico de Linhas**

```python
#Gráfico de Linha

plt.figure(figsize=(10, 7))

plt.plot(dataset['horario'], dataset['E'], linestyle='--', marker='o', color = 'blue', markersize = 4)
plt.title('Tempestade de 26/11/2011')
plt.legend(['Aeroporto'], fontsize=14)
plt.xlabel('Dia - horário', fontsize=15)
plt.ylabel('Campo Elétrico (V/m)', fontsize=15)

axes = plt.gca()
axes.yaxis.grid(b=True, color='black', alpha=0.3, linestyle='-.', linewidth=1)

plt.tight_layout()
plt.show()
```
